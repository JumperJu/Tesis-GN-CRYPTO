```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
from binance  import Client  #pip install python-binance
%run ./Binance_keys.ipynb
%load_ext sql
##%pip install SQLAlchemy --quiet
```

```python
client = Client(api_key,api_secret)
```

```python
engine = create_engine('sqlite:///Crypto_1h.db',echo=True)
```

```python
def gethourdata(symbol, start):
    frame = pd.DataFrame(client.get_historical_klines(symbol,
                                                     '1m',
                                                     start))
    frame = frame[[0,1,2,3,4]]
    frame.columns =['Date','Open','High','Low','Close']
    frame.Date = pd.to_datetime(frame.Date, unit='ms')
    frame.set_index('Date', inplace=True)
    frame = frame.astype(float)
    return frame
```

```python
#test
gethourdata('BTCUSDT','1 days ago UTC')
```

```python
#test coins
symbols = pd.read_csv('symbols.csv')
symbols
```

```python
#get data function
for symbol in symbols.name:
    df = gethourdata (symbol,'80 days ago UTC')
    df.to_sql(symbol, engine)
```
